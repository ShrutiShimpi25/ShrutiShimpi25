```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns .countplot(data=df , x='type', palette='pastel')
plt.title('count of movies and Tv shows on netflix')
plt.xlabel('content type')
plt.ylabel('count')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar', color='orange')
plt.title('top 10 countries')
plt.xlabel('country')
plt.ylabel('number of titles')
plt.xticks(rotation=45)
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', color='orange')
plt.title('content relesed over the year')
plt.xlabel('year')
plt.ylabel('number of titles')
plt.xticks(rotation=90)
plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)
top_directors.plot(kind='bar', color='purple')
plt.title('top 10 director')

plt.xlabel('number of titles')
plt.gca().invert_yaxis()
plt.show()
```
